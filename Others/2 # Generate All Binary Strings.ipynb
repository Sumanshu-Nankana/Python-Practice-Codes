```python
def generate_all_binary_strings(n):
    list_0 = []
    list_1 = []
    if n == 0: return []
    if n == 1: return ['0', '1']
    else:
        for i in generate_all_binary_strings(n-1):
            list_0 = list_0 + ['0'+i]
        for i in generate_all_binary_strings(n-1):
            list_1 = list_1 + ['1'+i]
        return list_0 + list_1
    
n = int(input("Enter any number: "))
out = generate_all_binary_strings(n)
print(f"Here is the all {n} digits possible binary strings: {out}")
```
